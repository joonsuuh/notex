```python
# 400 coinflips with 0.5 probability repeated 1`000`000 times and plot the probability distribution
n = 400
p = 0.5
size = 1000000
x = np.random.binomial(n, p, size)
plt.hist(x, bins=100)
plt.legend()
plt.show()

```
